```python
total_income = 0
total_expense = 0

while True:

    choice = input("Enter type (income/expense) or 'exit' to finish: ").lower()

    if choice == "income":
        amount = int(input("Enter income amount: "))
        total_income += amount
        print("Income added successfully!")

    elif choice == "expense":
        amount = int(input("Enter expense amount: "))
        total_expense += amount
        print("Expense added successfully!")

    elif choice == "exit":
        break

    else:
        print("Invalid input! Please type income, expense, or exit.")
        continue

    # Calculate net savings
    net_savings = total_income - total_expense

    print("--CURRENT SUMMARY--")
    print("Total Income:", total_income)
    print("Total Expenses:", total_expense)
    print("Net Savings:", net_savings)

    # Add insufficient balance condition
    if net_savings < 500:
        print("Insufficient balance!")



# Final summary
print("--FINAL SUMMARY--")
print("Total Income:", total_income)
print("Total Expenses:", total_expense)
print("Net Savings:", total_income - total_expense)

if (total_income - total_expense) < 500:
    print("Insufficient balance!")
```
